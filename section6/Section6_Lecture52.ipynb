## LINE CHART
### Line graphs are used to track changes over short and long periods of time by plotting data
### points and connecting them with a line

```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
df2 = pd.read_csv("happy2015.csv")
df2.head(5)
```

```python
df3=df2[["Happiness Score","Economy (GDP per Capita)","Family","Freedom","Generosity"]]
```

```python
df3.head(5)
```

```python
#plot a numerical value as line chart
```

```python
plt.plot(df3['Happiness Score'])
plt.ylabel('Happiness Score')
plt.xlabel('')
plt.show()
 
```

```python
plt.plot(df3['Happiness Score'],df3['Family'])
plt.ylabel('Family')
plt.xlabel('Happiness Score')
plt.title("Happiness & Family")
plt.show()
#relationship bw happiness core and family
```

#### Time Series of rainfall data from the different subdivisions of India (1901-2015)

```python
rain1=pd.read_csv("rainIndia.csv")
```

```python
rain1.head(4)
```

```python
rain1.SUBDIVISION.unique() #list the sub-divisions in India
```

```python
andnico=rain1[rain1.SUBDIVISION == "ANDAMAN & NICOBAR ISLANDS"]
```

```python
andnico.head(6)
```

```python
plt.plot(andnico['YEAR'],andnico['ANNUAL'])
plt.ylabel('Annual Rainfall')
plt.xlabel('Year')
plt.title("Andaman & Nicobar Rainfall Time Series")
plt.show()
```

```python
x=rain1.loc[rain1['SUBDIVISION'].isin(['ANDAMAN & NICOBAR ISLANDS','CHHATTISGARH','TAMIL NADU','TELANGANA','EAST RAJASTHAN'])]
```

```python
x.tail(6)
```

```python
x.SUBDIVISION.unique()
```

```python
plt.plot(x['YEAR'],x['ANNUAL'])
plt.ylabel('Annual Rainfall')
plt.xlabel('Year')
plt.title("India Rainfall Time Series")
plt.show()
```

```python
annico=x[x.SUBDIVISION == "ANDAMAN & NICOBAR ISLANDS"]
ch=x[x.SUBDIVISION == "CHHATTISGARH"]
er=x[x.SUBDIVISION == "EAST RAJASTHAN"]
tn=x[x.SUBDIVISION == "TAMIL NADU"]
t=x[x.SUBDIVISION == "TELANGANA"]
```

```python
fig = plt.figure(figsize=(15, 15)) #size of plot
plt.plot(annico['YEAR'],annico['ANNUAL'], label='Andaman')
plt.plot(ch['YEAR'],ch['ANNUAL'], marker='o', linestyle='--', color='r', label='Chattisgarh')
plt.plot(er['YEAR'],er['ANNUAL'], marker='o', linestyle='--', color='g', label='E Rajasthan')
plt.plot(tn['YEAR'],tn['ANNUAL'], marker='o', linestyle='--', color='y', label='TN')

plt.ylabel('Annual Rainfall')
plt.xlabel('Year')
plt.title("India Rainfall Time Series")
plt.legend()
plt.show()
```
